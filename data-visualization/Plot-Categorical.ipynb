```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Categorical Plot dengan Matplotlib

### Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
ax.bar(langs,students)
plt.show()
```

```python
data = [
        [30, 25, 50, 20],
        [40, 23, 51, 17],
        [35, 22, 45, 19]
       ]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(X+0.00, data[0], color='b', width=0.25)
ax.bar(X+0.25, data[1], color='g', width=0.25)
ax.bar(X+0.50, data[2], color='r', width=0.25)
```

# Categorical Plot dengan Seaborn

## Barplots

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

Kita juga dapat menambahkan dimensi pada data yang divisualisasikan dengan bar plot, yaitu dengan cara menambahkan parameter `hue`.

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips)
```

Menambahkan gradasi warna dengan parameter `palette`. Lebih lengkap mengenai `palette` pada seaborn dapat langsung mengunjungi [API Docs](https://seaborn.pydata.org/tutorial/color_palettes.html) resmi Seaborn.

```python
sns.barplot(x='size', y='total_bill', data=tips, palette='Blues_d')
```

## Categorical scatterplots

```python
# Load Dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

Parameter `jitter` untuk mengendalikan kerenggangan data. Jika `jitter=False` maka scatter plot akan lebih berhimpitan.

```python
sns.catplot(x='day', y='total_bill', jitter=False, data=tips)
```

Pendekatan selanjutnya adalah menggunakan parameter `kind` yang di-set ke nilai `swarm`. Parameter ini akan membuat representasi data lebih baik, namun hanya cocok untuk jumlah data yang relatif sedikit.

```python
sns.catplot(x='day', y='total_bill', kind='swarm', data=tips)
```

Kita juga dapat menambahkan dimensi pada data yang divisualisasikan dengan scatter plot, yaitu dengan cara menambahkan parameter `hue` pada method `catplot()`.

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=tips)
```

## Studi Kasus

```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/pokemon.csv')
df.head()
```

### Bar Plot

```python
type_1 = df['Type 1'].unique()
type_1_num = df['Type 1'].value_counts()
```

```python
# Bar plot dengan Matplotlib
plt.bar(type_1, type_1_num)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Bar plot pada library Pandas
df['Type 1'].value_counts().plot(kind='bar')
plt.show()
```

```python
generation = df['Generation'].unique()
generation_num = df['Generation'].value_counts()
```

```python
# Bar plot dengan Matplotlib
plt.bar(generation, generation_num)
plt.show()
```

```python
# Bar plot pada library Pandas
df['Generation'].value_counts().plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

### Bar Plot Lebih dari Dua Dimensi

```python
x = df[['Total', 'HP', 'Generation']]
x.head()
```

```python
x.plot.scatter('Total', 'HP', c='Generation', colormap='viridis')
plt.show()
```

```python
sns.catplot(x='Generation', y='Total', hue='Legendary', data=df)
```

```python
sns.catplot(x='Generation', y='Total', hue='Legendary', kind='swarm', data=df)
```
