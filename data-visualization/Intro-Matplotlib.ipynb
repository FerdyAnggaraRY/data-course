## Tips Matplotlib Umum

Sebelum kita menggunakan Matplotlib lebih lanjut, sebaiknya kita mempelajari terlebih dahulu beberapa tips menggunakan Matplotlib pada Python.

### Import Modul Matplotlib

````python
import matplotlib as mpl
import matplotlib.pyplot as plt
````

Kita akan menggunakan interface ``plt`` diseluruh kode program, namun pada dasarnya kita dapat menggunakan interface tersendiri sesuai dengan keinginan kita.

#### Plot dari IPython shell

Untuk menggunakan konfigurasi Matplotlib yang disesuaikan dengan IPython shell kita bisa mencantumkan ``%matplotlib`` sebelum meng-import module Matplotlib.

```ipython
In [1]: %matplotlib
Using matplotlib backend: TkAgg

In [2]: import matplotlib.pyplot as plt
```

````python
%matplotlib inline
````

Setelah menjalankan kode program di atas, maka setiap gambar ditampilkan dalam format PNG secara otomatis.

````python
import numpy as np
x = np.linspace(0, 10, 100)
print(x)
````

Format untuk plotting `(sumbu x, sumbu y)`

````python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'bold',
        'size': 16,
        }
fig = plt.figure()
plt.plot(x, np.sin(x), '-') 
plt.plot(x, np.cos(x), '--');
plt.xlabel('Nilai x', fontdict=font)
plt.ylabel('Nilai Sin/Cos', fontdict=font)
````

````python
np.cos(x)
````

### Save Plot/Gambar ke File

````python
fig.savefig('my_figure.png')
````

## Dua Plot dalam Satu Shell Output

Berikut adalah contoh menampilkan dua plot dalam satu shell output.

|Template Subplot     |                     |
|---------------------|---------------------|
|Row 1, Col 1, Panel 1|Row 1, Col 2, Panel 3|
|Row 2, Col 1, Panel 2|Row 2, Col 2, Panel 4|

**``subplot(rows, colums, panel number)``**

````python
plt.figure()  # Membuat plot figure

# Menampilkan dua plot dengan format dua baris, satu kolom
plt.subplot(1, 2, 1) # Panel 1
plt.plot(x, np.sin(x))

plt.subplot(1, 2, 2) # Panel 2
plt.plot(x, np.cos(x));
````

#### Cara Kedua

````python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
````

### Set Ukuran Gambar

Ukuran gambar diimplementasikan dengan method `figure` dengan parameter `figsize`. Format data untuk parameter `figsize` adalah python dictionary `(sumbu x, sumbu y)`, di mana `x` dan `y` dalam ukuran inchi.

````python
import numpy as np

x = np.linspace(0, 10, 100)
plt.figure(figsize=(20, 12))
plt.plot(x, np.cos(x))
plt.show()
````
